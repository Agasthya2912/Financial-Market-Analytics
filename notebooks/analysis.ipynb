```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/apple_features.csv")

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'])
plt.title("Apple Stock Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Date'], df['Close'], label="Close Price")
plt.plot(df['Date'], df['MA_10'], label="10 Day MA")
plt.plot(df['Date'], df['MA_50'], label="50 Day MA")

plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# download multiple stocks
stocks = ["AAPL", "TSLA", "MSFT", "GOOGL"]
data = yf.download(stocks, start="2018-01-01", end="2024-01-01")["Close"]

# calculate returns
returns = data.pct_change().dropna()

# correlation matrix
corr_matrix = returns.corr()

# heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Stock Correlation Heatmap (Tech Stocks)")
plt.show()
```
